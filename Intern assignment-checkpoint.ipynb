```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# reading dataset to python environment
data1=pd.read_excel("asgnmnt.xlsx")
data2=pd.read_excel("asgmnt2.xlsx")
```

```python
# dispalying dataset
data1
```

```python
# displaying dataset 
data2
```

```python
# concatenating datasets
df=pd.concat([data1,data2],join='outer',axis=1)
df.drop(['S No','Name','User ID'],axis=1, inplace=True)
df.rename(columns={'name':'Name','uid':'User ID'},inplace=True)
df
```

### Teamwise Leaderboard

```python
# saviing the dataset with a new name
df_team=df
df_team
```

```python
# group by data by team
team=df_team.groupby(['Team Name']).mean()
team.drop('User ID',axis=1,inplace=True)
team.reset_index(inplace=True)
```

```python
# renaming columns, sorting and setting the decimals
team.rename(columns={'Team Name':'Thinking Teams Leaderboard','total_statements':'Average Statements','total_reasons':'Average Reasons'},inplace=True)
team.sort_values(by=['Average Statements','Average Reasons'],ascending=[0,0],inplace=True)
team=team.round(decimals=2)
team
```

```python
# setting the Rank column
team.index = np.arange(1, len(team) + 1)
team.reset_index(inplace=True)
team.rename(columns={'index':'Rank'},inplace=True)
team.set_index('Rank',inplace=True)
team
```

```python
# saving the file
team.to_csv('D:internships/team.csv')# index =False is used to not to give index as a clmn
```
